```python
import pandas as pd
import random
from faker import Faker

# Create a Faker instance
fake = Faker()

# Create a dictionary with the data
data = {
    'ID': range(1, 201),
    'Date': pd.date_range(start='01-01-2023', periods=200),
    'Owner': [fake.name() for _ in range(200)],
    'Account': [fake.unique.company() for _ in range(200)],
    'Subscription costs': [round(random.uniform(100, 500), 2) for _ in range(200)],
    'Support costs': [round(random.uniform(50, 200), 2) for _ in range(200)],
    'Services costs': [round(random.uniform(75, 300), 2) for _ in range(200)],
    'Amount': [round(random.uniform(225, 750), 2) for _ in range(200)]
}

# Create a DataFrame
df = pd.DataFrame(data)

df.head()
```

```python
# Write DataFrame to Excel
df.to_excel('orders.xlsx', index=False)
```
