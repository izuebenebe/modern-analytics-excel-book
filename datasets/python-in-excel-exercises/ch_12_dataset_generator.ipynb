```python

Certainly! Let's create a synthetic dataset related to e-commerce data, specifically focusing on customer behavior and product reviews. Here's the modified script:

python
Copy code
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a random seed for reproducibility
np.random.seed(42)

# Generate a time series index
date_rng = pd.date_range(start='2020-01-01', end='2020-12-31', freq='D')

# Create a DataFrame with date index
df = pd.DataFrame(date_rng, columns=['date'])

# Generate numerical features
num_features = ['product_price', 'quantity_purchased', 'customer_rating']
for feature in num_features:
    df[feature] = np.random.uniform(10, 500, size=len(date_rng))

# Generate categorical features
categories = ['product_category', 'customer_segment']
product_categories = ['Electronics', 'Clothing', 'Books', 'Home Appliances']
customer_segments = ['Regular', 'Premium', 'VIP']

df['product_category'] = np.random.choice(product_categories, size=len(date_rng))
df['customer_segment'] = np.random.choice(customer_segments, size=len(date_rng))

# Introduce missing values
for feature in num_features + categories:
    missing_mask = np.random.rand(len(date_rng)) < 0.2  # 20% missing values
    df.loc[missing_mask, feature] = np.nan


df.to_excel('ecommerce.xlsx', Index=False)
```
