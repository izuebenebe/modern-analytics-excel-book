The objective of this notebook is to create a brief Excel report using `openpyxl`. 

```python
import pandas as pd
import seaborn as sns
df = pd.read_excel('contestants_cleaned.xlsx')
```

# Create a summary worksheet

For the first step, we are going to print the average pre and post test scores on a summary worksheet. 

This is of course what. I'm not sure. 

```python
from openpyxl import Workbook

wb = Workbook()
ws = wb.active
ws.title = 'summary'
```

We are going to hard-code some summary labels up in here 

```python
# There are a couple of ways to do this! 
ws['A1'] = 'Average pre score:'
ws.cell(row=2, column=1, value='Average post score:')
```

We are now going to plop in the calculated values next to these, yo!

```python
pre_mean = round(df['pre'].mean(), 1)
print(pre_mean)
ws['B1'] = pre_mean
```

```python
post_mean = round(df['post'].mean(), 1)
print(post_mean)
ws['B2'] = post_mean
```
